# Transformación para el set de testeo 

Con este dataset haré las predicciones que enviaré al datathon para que se evaluado. 

```
import pandas as pd
import numpy as np
```

```
# Carga de datos
testeo = pd.read_parquet(r'test.parquet')
```

## Exploración de dataset

Estos datos no contienen la columna price.

```
# visualizo los primeros 5 registros de mi dataframe
testeo.head()
```

En este caso aplicaré al dataset de testeo (train.parquet) todos los cambios que realicé al test de entrenamiento (train.parquet), sin embargo solo los cambios a nivel de columnas, ya que si aplicara cambios a nivel de registros (filas) los datos predictorios serían inconsistentes al momento de enfrentarlos con los datos reales en su testeo. 

```
testeo.drop('laundry_options',axis=1,inplace=True)
testeo.drop('parking_options',axis=1,inplace=True)
```

```
testeo.drop('id', axis=1, inplace=True)
testeo.drop('url', axis=1, inplace=True)
testeo.drop('region_url', axis=1, inplace=True)
testeo.drop('image_url', axis=1, inplace=True)
testeo.drop('description',axis=1, inplace=True)
testeo.drop('region', axis=1, inplace=True)
testeo.drop('state', axis=1, inplace=True)

```

```
# Reviso que los datos hayan sido aplicados correctamente. 
testeo
```

```
# Transormo el dataset obteniendo las variables dummies. 
testeo = pd.get_dummies(testeo)
```

```
# Rectifico los cambios
testeo
```

```
# Rectifico que el número de registros coincidan con el número de valores para la evaluación del modelo
# por parte del datathon. 
testeo.shape
```

```
# Finalmente obtengo mi archivo csv
testeo.to_csv('testeo.csv', sep=',', index=False)
```
